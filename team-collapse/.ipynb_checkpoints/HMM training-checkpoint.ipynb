```python
import numpy as np
import pickle
from hmmlearn import hmm
import cvxpy
```

```python
n_components = 3 # number of hidden states
initial_dist = np.array([0, 1, 0])
with open('./observations/BOS_observations.pickle', 'rb') as file:
    observations = pickle.load(file)
```

```python
observations
```

```python
players = observations[list(observations.keys())[0]].keys()
players
```

```python
window_size = len(observations[list(observations.keys())[0]][list(players)[0]])
window_size
```

```python
H = dict()
O = dict()
for player in players:
    model = hmm.CategoricalHMM(n_components=n_components, init_params='te', params='te')
    model.startprob_ = initial_dist
    
    X = []
    for game_id in observations:
        X.append(observations[game_id][player])
    
    
    X = np.array(X)
    X = X.flatten()
    O[player] = X # save observations
    X = X.reshape(-1, 1)
    
    lengths = [window_size] * len(observations)
    
    
    model.fit(X, lengths=lengths)
    score = model.score(X, lengths=lengths)
    opt_model = model
    opt_score = score
    
#     for _ in range(10):
#         model.fit(X, lengths=lengths)
#         score = model.score(X, lengths=lengths)
#         if score > opt_score:
#             print('test')
#             opt_model = model
#             opt_score = score
    
    print(opt_score)
    
    H[player] = opt_model.predict(X, lengths=lengths) # MAP estimation of hidden states
    
    
```

```python
n_features = model.n_features # num of observed symbols
n_features
```

```python
M = dict() # stores transition probabilities from previous observations
N = dict() # stores transition probabilities from previous hidden state

for player1 in players:
    print(f'Player = {player1}')
    M[player1] = np.zeros((n_components, n_features)) # M[i][j] = Pr(H_t^player1 = i | O_{t-1}^player2 = j), for any player2
    N[player1] = np.zeros((n_components, n_components)) # N[i][j] = Pr(H_t^player1 = i | H_{t-1}^player1 = j) 
    
    hidden_states = H[player1]
    
    # Compute N
    L = list(zip(hidden_states[1:], hidden_states))
    for (h1, h2) in L: # h1 = next state, h2 = previous state
        N[player1][h1][h2] += 1

    N[player1] = N[player1] / N[player1].sum(axis=0, keepdims=True)
    
    # Compute M
    for player2 in players:
        observed_states = O[player2]
        
        L = list(zip(hidden_states[1:], observed_states))
        for (h, o) in L: # h = next state, o = previous observation
            M[player1][h][o] += 1
            
    M[player1] = M[player1] / M[player1].sum(axis=0, keepdims=True)
```
